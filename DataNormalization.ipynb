```python
df = cleaned_df.copy()


def normalize_sensor_data(df):
    df = df.copy()

    # Clip values to realistic ranges
    df['methane_pct'] = df['methane_pct'].clip(0, 2.5)
    df['temperature_c'] = df['temperature_c'].clip(15, 60)
    df['humidity_pct'] = df['humidity_pct'].clip(0, 100)
    df['vibration_mm_s'] = df['vibration_mm_s'].clip(0, 10)
    df['airflow_mps'] = df['airflow_mps'].clip(0, 6)
    df['oxygen_pct'] = df['oxygen_pct'].clip(18, 23)

    return df

```
